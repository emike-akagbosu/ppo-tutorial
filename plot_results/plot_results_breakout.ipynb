```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

filepath = r"C:\Users\emike\OneDrive - Imperial College London\Final Year\Reinforcement Learning\Coursework 2\Breakout\monitor_default.csv"
df = pd.read_csv(filepath)

rewards = df['r']
episode = df['episode']
episode_length = df['l']
```

```python
filepath2 = r"C:\Users\emike\OneDrive - Imperial College London\Final Year\Reinforcement Learning\Coursework 2\Breakout\monitor_hparams.csv"
df_params = pd.read_csv(filepath2)

rewards_params = df_params['r']
episode_params = df_params['episode']
episode_length_params = df_params['l']
```

```python
# plotting a line graph 
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.plot(episode, rewards, linewidth=2) 

plt.title('Mountain Car Continuous - Learning Curve', fontsize=16)
plt.xlabel('Episode', fontsize=12)
plt.ylabel('Rewards', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=1)

#plt.savefig("MCC-learning-curve-default-params-dark", dpi=300, bbox_inches='tight')

plt.show() 
```

```python
# plotting a line graph 
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.plot(episode, episode_length, linewidth=2) 

plt.title('Mountain Car Continuous - Learning Curve', fontsize=16)
plt.xlabel('Episode', fontsize=12)
plt.ylabel('Episode Length', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=1)

#plt.savefig("MCC-episode-length-default-params", dpi=300, bbox_inches='tight')

plt.show() 
```

```python
# plotting a line graph 
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.plot(episode, rewards, linewidth=2, label='Default Hyperparameters') 
plt.plot(episode_params, rewards_params, label='Decaying Entropy', color='green')

plt.title('Breakout - Learning Curve', fontsize=16)
plt.xlabel('Episode', fontsize=12)
plt.ylabel('Rewards', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=1)
plt.legend()

#plt.savefig("Breakout-rewards-both", dpi=300, bbox_inches='tight')

plt.show() 
```

```python
# plotting a line graph 
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.plot(episode, episode_length, linewidth=2, label='Default Hyperparameters') 
plt.plot(episode_params, episode_length_params, label='Decaying Entropy', color='green')

plt.title('Breakout - Learning Curve', fontsize=16)
plt.xlabel('Episode', fontsize=12)
plt.ylabel('Episode Length', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=1)
plt.legend()

#plt.savefig("Breakout-episode-length-both", dpi=300, bbox_inches='tight')

plt.show() 
```
