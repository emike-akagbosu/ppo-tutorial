```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

filepath = r"C:\Users\emike\OneDrive - Imperial College London\Final Year\Reinforcement Learning\Coursework 2\Breakout\mean_data.csv"
df = pd.read_csv(filepath)

mean_reward = df['mean reward']
timestep = df['Step']
mean_episode_length = df['mean episode length']
```

```python
filepath2 = r"C:\Users\emike\OneDrive - Imperial College London\Final Year\Reinforcement Learning\Coursework 2\Breakout\mean_data_hparams.csv"
df2 = pd.read_csv(filepath2)
df2 = df2.dropna()

mean_reward2 = df2['mean reward']
timestep2 = df2['Step']
mean_episode_length2 = df2['mean episode length']
```

```python
print(mean_reward2)
```

```python
# plotting a line graph 
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.plot(timestep, mean_reward, linewidth=2, label='Default Hyperparameters') 
plt.plot(timestep2, mean_reward2, label='Decaying Entropy', color='green')

plt.title('Breakout - Mean Rewards During Training', fontsize=16)
plt.xlabel('Timesteps', fontsize=12)
plt.ylabel('Rewards', fontsize=12)
#plt.xticks(np.arange(0, 1001, 100))
#plt.yticks(np.arange(-60, 10, 10))
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=1)
plt.legend()

plt.savefig("Breakout-rewards-mean", dpi=300, bbox_inches='tight')

plt.show() 
```

```python
# plotting a line graph 
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.plot(timestep, mean_episode_length, linewidth=2, label='Default Hyperparameters') 
plt.plot(timestep2, mean_episode_length2, label='Decaying Entropy', color='green')

plt.title('Breakout - Mean Episode Length During Training', fontsize=16)
plt.xlabel('Timesteps', fontsize=12)
plt.ylabel('Episode Length', fontsize=12)
#plt.xticks(np.arange(0, 1001, 100))
#plt.yticks(np.arange(-60, 10, 10))
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=1)
plt.legend()

#plt.savefig("Breakout-episode-length-mean2", dpi=300, bbox_inches='tight')

plt.show() 
```
